*   å¦‚æœæ‚¨æœ‰å¤§è§„æ¨¡æ•°æ®ï¼Œå¯ä»¥åœ¨ Docker æˆ– Kubernetes ä¸Šè®¾ç½®æ€§èƒ½æ›´é«˜çš„ Milvus æœåŠ¡å™¨ã€‚åœ¨æ­¤è®¾ç½®ä¸­ï¼Œè¯·ä½¿ç”¨æœåŠ¡å™¨ URIï¼Œä¾‹å¦‚ `http://localhost:19530`ï¼Œä½œä¸ºæ‚¨çš„ `uri`ã€‚

æ£€æŸ¥ collection æ˜¯å¦å·²å­˜åœ¨ï¼Œå¦‚æœå­˜åœ¨åˆ™åˆ é™¤å®ƒã€‚**è¿™æ˜¯ä¸ºäº†ç¤ºä¾‹ä»£ç å¯ä»¥é‡å¤è¿è¡Œ**

```python
if milvus_client.has_collection(collection_name):
    milvus_client.drop_collection(collection_name)
```

åˆ›å»ºä¸€ä¸ªå…·æœ‰æŒ‡å®šå‚æ•°çš„æ–° collectionã€‚

å¦‚æœæˆ‘ä»¬ä¸æŒ‡å®šä»»ä½•å­—æ®µä¿¡æ¯ï¼ŒMilvus å°†è‡ªåŠ¨åˆ›å»ºä¸€ä¸ªé»˜è®¤çš„ `id` å­—æ®µä½œä¸ºä¸»é”®ï¼Œä»¥åŠä¸€ä¸ª `vector` å­—æ®µæ¥å­˜å‚¨å‘é‡æ•°æ®ã€‚ä¸€ä¸ªä¿ç•™çš„ JSON å­—æ®µç”¨äºå­˜å‚¨é schema å®šä¹‰çš„å­—æ®µåŠå…¶å€¼ã€‚

`metric_type` (è·ç¦»åº¦é‡ç±»å‹):
     ä½œç”¨ï¼šå®šä¹‰å¦‚ä½•è®¡ç®—å‘é‡ä¹‹é—´çš„ç›¸ä¼¼ç¨‹åº¦ã€‚
     ä¾‹å¦‚ï¼š`IP` (å†…ç§¯) - å€¼è¶Šå¤§é€šå¸¸è¶Šç›¸ä¼¼ï¼›`L2` (æ¬§æ°è·ç¦») - å€¼è¶Šå°è¶Šç›¸ä¼¼ï¼›`COSINE` (ä½™å¼¦ç›¸ä¼¼åº¦) - é€šå¸¸è½¬æ¢ä¸ºè·ç¦»ï¼Œå€¼è¶Šå°è¶Šç›¸ä¼¼ã€‚
     é€‰æ‹©ä¾æ®ï¼šæ ¹æ®ä½ çš„åµŒå…¥æ¨¡å‹çš„ç‰¹æ€§å’ŒæœŸæœ›çš„ç›¸ä¼¼æ€§å®šä¹‰æ¥é€‰æ‹©ã€‚

 `consistency_level` (ä¸€è‡´æ€§çº§åˆ«):
     ä½œç”¨ï¼šå®šä¹‰æ•°æ®å†™å…¥åï¼Œè¯»å–æ“ä½œèƒ½å¤šå¿«çœ‹åˆ°è¿™äº›æ–°æ•°æ®ã€‚
     ä¾‹å¦‚ï¼š
         `Strong` (å¼ºä¸€è‡´æ€§): æ€»æ˜¯è¯»åˆ°æœ€æ–°æ•°æ®ï¼Œå¯èƒ½ç¨æ…¢ã€‚
         `Bounded` (æœ‰ç•Œè¿‡æœŸ): å¯èƒ½è¯»åˆ°å‡ ç§’å†…æ—§æ•°æ®ï¼Œæ€§èƒ½è¾ƒå¥½ (é»˜è®¤)ã€‚
         `Session` (ä¼šè¯ä¸€è‡´æ€§): è‡ªå·±å†™å…¥çš„è‡ªå·±èƒ½ç«‹åˆ»è¯»åˆ°ã€‚
         `Eventually` (æœ€ç»ˆä¸€è‡´æ€§): æœ€ç»ˆä¼šè¯»åˆ°æ–°æ•°æ®ï¼Œä½†æ²¡æ—¶é—´ä¿è¯ï¼Œæ€§èƒ½æœ€å¥½ã€‚
     é€‰æ‹©ä¾æ®ï¼šåœ¨æ•°æ®å®æ—¶æ€§è¦æ±‚å’Œç³»ç»Ÿæ€§èƒ½ä¹‹é—´åšæƒè¡¡ã€‚

ç®€å•æ¥è¯´ï¼š
 `metric_type`ï¼šæ€ä¹ˆç®—ç›¸ä¼¼ã€‚
 `consistency_level`ï¼šæ–°æ•°æ®å¤šä¹…èƒ½è¢«è¯»åˆ°ã€‚

```python
milvus_client.create_collection(
    collection_name=collection_name,
    dimension=embedding_dim,
    metric_type="IP",  # å†…ç§¯è·ç¦»
    consistency_level="Strong",  # æ”¯æŒçš„å€¼ä¸º (`"Strong"`, `"Session"`, `"Bounded"`, `"Eventually"`)ã€‚æ›´å¤šè¯¦æƒ…è¯·å‚è§ https://milvus.io/docs/consistency.md#Consistency-Levelã€‚
)
```

### æ’å…¥æ•°æ®

éå†æ–‡æœ¬è¡Œï¼Œåˆ›å»ºåµŒå…¥ï¼Œç„¶åå°†æ•°æ®æ’å…¥ Milvusã€‚

è¿™é‡Œæœ‰ä¸€ä¸ªæ–°å­—æ®µ `text`ï¼Œå®ƒæ˜¯åœ¨ collection schema ä¸­æœªå®šä¹‰çš„å­—æ®µã€‚å®ƒå°†è‡ªåŠ¨æ·»åŠ åˆ°ä¿ç•™çš„ JSON åŠ¨æ€å­—æ®µä¸­ï¼Œè¯¥å­—æ®µåœ¨é«˜çº§åˆ«ä¸Šå¯ä»¥è¢«è§†ä¸ºæ™®é€šå­—æ®µã€‚

```python
from tqdm import tqdm

data = []

doc_embeddings = embedding_model.encode_documents(text_lines)

for i, line in enumerate(tqdm(text_lines, desc="Creating embeddings")):
    data.append({"id": i, "vector": doc_embeddings[i], "text": line})

milvus_client.insert(collection_name=collection_name, data=data)
```

### æ£€ç´¢æŸ¥è¯¢æ•°æ®

æˆ‘ä»¬æŒ‡å®šä¸€ä¸ªå…³äº Milvus çš„å¸¸è§é—®é¢˜ã€‚

```python
question = "How is data stored in milvus?"
```

åœ¨ collection ä¸­æœç´¢è¯¥é—®é¢˜ï¼Œå¹¶æ£€ç´¢è¯­ä¹‰ä¸Šæœ€åŒ¹é…çš„å‰3ä¸ªç»“æœã€‚

```python
search_res = milvus_client.search(
    collection_name=collection_name,
    data=embedding_model.encode_queries(
        [question]
    ),  # å°†é—®é¢˜è½¬æ¢ä¸ºåµŒå…¥å‘é‡
    limit=3,  # è¿”å›å‰3ä¸ªç»“æœ
    search_params={"metric_type": "IP", "params": {}},  # å†…ç§¯è·ç¦»
    output_fields=["text"],  # è¿”å› text å­—æ®µ
)
```

è®©æˆ‘ä»¬çœ‹ä¸€ä¸‹æŸ¥è¯¢çš„æœç´¢ç»“æœ

```python
import json

retrieved_lines_with_distances = [
    (res["entity"]["text"], res["distance"]) for res in search_res[0]
]
print(json.dumps(retrieved_lines_with_distances, indent=4))
```
